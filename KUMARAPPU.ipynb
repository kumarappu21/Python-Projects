```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
delattr = pd.read_excel(r'C:\Users\KUMAR APPU\Downloads\onlineretail.xlsx')
delattr.head()
```

```python
delattr.tail()
```

```python
delattr = delattr.loc[delattr['Quantity'] > 0]
```

```python
delattr.info()
```

```python
delattr['CustomerID'].isna().sum()
delattr = delattr.dropna(subset=['CustomerID'])
```

```python
customer_item_matrix = delattr.pivot_table(
    index='CustomerID',
    columns='StockCode',
    values='Quantity',
    aggfunc='sum'
)
customer_item_matrix.loc[17850.0:].head()
```

```python
print(customer_item_matrix.shape)
customer_item_matrix = customer_item_matrix.applymap(lambda x: 1 if x > 0 else 0)
```

```python
user_user_sim_matrix = pd.DataFrame(cosine_similarity(customer_item_matrix))
user_user_sim_matrix.head()
```

```python
user_user_sim_matrix.columns = customer_item_matrix.index

user_user_sim_matrix['CustomerID'] = customer_item_matrix.index
user_user_sim_matrix = user_user_sim_matrix.set_index('CustomerID')
user_user_sim_matrix.head()
```

```python
user_user_sim_matrix.loc[12346.0].sort_values(ascending=False).head(10)
```

```python
user_user_sim_matrix.loc[12346.0].sort_values(ascending=False)
items_bought_by_A = customer_item_matrix.loc[12346.0][customer_item_matrix.loc[12346.0]>0]
print("Items Bought by A: ")
print(items_bought_by_A)
```

```python
items_bought_by_B = customer_item_matrix.loc[17850.0][customer_item_matrix.loc[17850.0]>0]
print("Items bought by B:")
print(items_bought_by_B)

print()

items_to_recommend_to_B = set(items_bought_by_A.index) - set(items_bought_by_B.index)
print("Items to Recommend to B ")
print(items_to_recommend_to_B)
delattr.loc[delattr['StockCode'].isin(items_to_recommend_to_B),['StockCode', 'Description']].drop_duplicates().set_index('StockCode')
```

```python
item_item_sim_matrix = pd.DataFrame(cosine_similarity(customer_item_matrix.T))
item_item_sim_matrix.columns = customer_item_matrix.T.index

item_item_sim_matrix['StockCode'] = customer_item_matrix.T.index
item_item_sim_matrix = item_item_sim_matrix.set_index('StockCode')
```
